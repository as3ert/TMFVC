### Exercise 9.1 (8 points)
According to the lecture, the Fourier series of a function $f: [-\pi,\pi] \rightarrow \mathbb{R}$ is given by
\begin{equation} 
    g(x) = \frac{1}{2} a_0 + \sum_{k=1}^\infty \big[ a_k \cos(kx) + b_k \sin(kx) \big],
\end{equation}
where $a_0, a_1, a_2, \ldots \in \mathbb{R}$ and $b_1, b_2, \ldots \in \mathbb{R}$ are given by the formula
\begin{align*}
    a_k &= \frac{1}{\pi} \int_{-\pi}^{\pi} f(x) \cos(kx) ~dx, \\
    b_k &= \frac{1}{\pi} \int_{-\pi}^{\pi} f(x) \sin(kx) ~dx.
\end{align*}
The Fourier series can be used to represent/approximate a function via a sum of simple trigonometric functions.
For this, the first Equation (1) is rearranged into a finite sum (Fourier polynomial):
\begin{equation*}
    g_n(x) = \frac{1}{2} a_0 + \sum_{k=1}^n \big[ a_k \cos(kx) + b_k \sin(kx) \big].
\end{equation*}

Explain both concepts and their relationship in your own words and discuss the keywords *time* and *time domain* as well as *frequency* and *frequency domain*.

**Answer:**

- Fourier Transform:
  - Fourier transform converts continous spatial (or time) domain to frequency domain.
- Fourier Series:
  - Fourier series converts a periodic function to a sum of trigonometric functions.
- Relationship:
  - Fourier transform can handel non-periodic signals, which is an extention of Fourier series.

- Time and Time Domain:
  - Time refers to the progression over time.
  - Time domain is the domain where the signal's value changes along with time.

- Frequency and Frequency Domain:
  - Frequency refers to the number of cycles of a periodic signal in a unit time.
  - Frequency domain is the domain where the signal's amplitude changes along with frequency.


`(b)`Given the periodically continued function $f: [-\pi, \pi) \rightarrow \mathbb{R}$ with
\begin{equation}
    f(x) = \begin{cases}
        1, \qquad \qquad x\in[-\pi, 0), \\
        1 - x/\pi, \quad \, x\in[0,\pi),
    \end{cases}
\end{equation}
complete the following tasks:
1. Calculate the coefficients $a_k, b_k$ of the Fourier polynomial $g_{n}(x)$ of $f$ dependent on $k$.
2. Write a Python script to plot $f$ together with $g_3$, $g_{10}$, $g_{25}$, and $g_{100}$ for $x\in [-3\pi, 3\pi]$. 
>Hint: The function $f$ is periodically continued.

**Answer:**

According to the Fourier polynomial formula:

\begin{align*}
    a_k &= \frac{1}{\pi} \int_{-\pi}^{\pi} f(x) \cos(kx) \, \mathrm{d}x, \\
    b_k &= \frac{1}{\pi} \int_{-\pi}^{\pi} f(x) \sin(kx) \, \mathrm{d}x, \\
    g_{n}(x) &= \frac{1}{2} a_0 + \sum_{k=1}^n \big[ a_k \cos(kx) + b_k \sin(kx) \big],
\end{align*}

we can calculate the coefficients $a_k, b_k$ of the Fourier polynomial $g_{n}(x)$ of $f$ dependent on $k$ as follows:

\begin{align*}
    a_k &= \frac{1}{\pi} \int_{-\pi}^{0} \cos(kx) \, \mathrm{d}x + \frac{1}{\pi} \int_{0}^{\pi} (1-\frac{x}{\pi}) \cos(kx) \, \mathrm{d}x, \\
        &= \frac{1}{\pi} \int_{-\pi}^{\pi} \cos(kx)  \, \mathrm{d}x - \frac{1}{\pi^2} \int_{0}^{\pi} x \cos(kx) \, \mathrm{d}x, \\
        &= -\frac{1}{\pi^2} \int_{0}^{\pi} x \cos(kx) \, \mathrm{d}x, \\
        &= -\frac{1}{\pi^2} \Big[\frac{x}{k} \sin(kx) + \frac{1}{k^2} \cos(kx) \Big]_{0}^{\pi}, \\
        &= -\frac{\cos(k\pi) - 1}{k^2\pi^2}.
\end{align*}

\begin{align*}
    b_k &= \frac{1}{\pi} \int_{-\pi}^{0} \sin(kx) \, \mathrm{d}x + \frac{1}{\pi} \int_{0}^{\pi} (1-\frac{x}{\pi}) \sin(kx) \, \mathrm{d}x, \\
        &= \frac{1}{\pi} \int_{-\pi}^{\pi} \sin(kx)  \, \mathrm{d}x - \frac{1}{\pi^2} \int_{0}^{\pi} x \sin(kx) \, \mathrm{d}x, \\
        &= -\frac{1}{\pi^2} \int_{0}^{\pi} x \sin(kx) \, \mathrm{d}x, \\
        &= -\frac{1}{\pi^2} \Big[-\frac{x}{k} \cos(kx) + \frac{1}{k^2} \sin(kx) \Big]_{0}^{\pi}, \\
        &= \frac{\cos(k\pi)}{k\pi}.
\end{align*}

For $a_0$, we have:

\begin{align*}
    a_0 &= \frac{1}{\pi} \int_{-\pi}^{0} 1 \, \mathrm{d}x + \frac{1}{\pi} \int_{0}^{\pi} (1-\frac{x}{\pi}) \, \mathrm{d}x, \\
        &= \frac{1}{\pi} \int_{-\pi}^{\pi} 1 \, \mathrm{d}x - \frac{1}{\pi^2} \int_{0}^{\pi} x \, \mathrm{d}x, \\
        &= 2 - \frac{1}{\pi^2} \Big[\frac{x^2}{2} \Big]_{0}^{\pi}, \\
        &= \frac{3}{2}.
\end{align*}

Thus, the Fourier polynomial $g_{n}(x)$ of $f$ dependent on $k$ is:

\begin{align*}
    g_{n}(x) &= \frac{1}{2} a_0 + \sum_{k=1}^n \big[ a_k \cos(kx) + b_k \sin(kx) \big], \\
        &= \frac{3}{4} + \sum_{k=1}^n \big[ -\frac{\cos(k\pi) - 1}{k^2\pi^2} \cos(kx) + \frac{\cos(k\pi)}{k\pi} \sin(kx) \big].
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt

K = [3, 10, 25, 100]
x_range = np.linspace(-3*np.pi, 3*np.pi, 1000)

def f(x):
    x = x % (2 * np.pi)

    if np.pi < x <= 2 * np.pi:
        return 1
    elif 0 <= x <= np.pi:
        return 1 - x / np.pi
    
f_vec = np.vectorize(f)

def g(n, x):
    sum = 3 / 4

    for i in range(1, n+1):
        sum += -(np.cos(i * np.pi) - 1) * np.cos(i * x) / (i * np.pi)**2 + np.cos(i * np.pi) * np.sin(i * x) / (i * np.pi)

    return sum

plt.figure(figsize=(10, 10))

for i in range(len(K)):
    plt.subplot(4, 1, i + 1)
    plt.plot(x_range, f_vec(x_range), label='$f(x)$', color='black', linewidth=2)
    plt.plot(x_range, g(K[i], x_range), label=f'$g_{{{K[i]}}}(x)$', linewidth=1.5)

    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.legend()
    plt.grid()
    plt.title(f'Fourier Approximation with n = {K[i]}', fontsize=12)

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

`(a)` Show that the Fourier transform of $f(x) = \sin(k_0x)$ is:

where $\delta$ is the Dirac delta. 

**Answer:**

For the Fourier transform of $f(x) = \sin(k_0x)$, we have:

\begin{align*}
    \mathcal{F}[f(x)](k) &= \int_{-\infty}^{+\infty} \sin(k_0x) e^{-ikx} \, \mathrm{d}x, \\
        &= \int_{-\infty}^{+\infty} \frac{1}{2i} \big[ \cos(k_0x) + i \sin(k_0x) - cos(k_0x) + i \sin(k_0x) \big] e^{-ikx} \, \mathrm{d}x, \\
        &= \frac{1}{2i} \int_{-\infty}^{+\infty} (e^{ik_0x} - e^{-ik_0x}) e^{-ikx} \, \mathrm{d}x, \\
        &= \frac{1}{2i} \int_{-\infty}^{+\infty} e^{ik_0x} e^{-ikx} \, \mathrm{d}x - \frac{1}{2i} \int_{-\infty}^{+\infty} e^{-ik_0x} e^{-ikx} \, \mathrm{d}x, \\
        &= \frac{1}{2i} \mathcal{F}[e^{ik_0x}](k) - \frac{1}{2i} \mathcal{F}[e^{-ik_0x}](k). \\
\end{align*}

Recall that $g(x)=e^{iax}, \mathcal{F}[g(x)](k) = \delta(k-a)$, we have:

\begin{align*}
    \mathcal{F}[f(x)](k) &= \frac{1}{2i} \mathcal{F}[e^{ik_0x}](k) - \frac{1}{2i} \mathcal{F}[e^{-ik_0x}](k), \\
        &= \frac{1}{2i}[\delta(k-k_0) - \delta(k+k_0)].
\end{align*}

`(b)` Show for the convolution $[F \star F](k)$:
\begin{equation*}
    [F \star F](k)=\mathcal{F}[g(x)g(x)](k)
\end{equation*}
where $F(k) = \mathcal{F}[g(x)](k)$ is the Fourier transform of $g(x)$, and $g(x)$ is an arbitrary analytic function. 

**Answer:**

For the convolution $[F \star F](k)$, we have:

\begin{align*}
    [F \star F](k) &= \int_{-\infty}^{+\infty} F(u) F(k-u) \, \mathrm{d}u, \\
        &= \int_{-\infty}^{+\infty} \mathcal{F}[g(x)](u) \mathcal{F}[g(x)](k-u) \, \mathrm{d}u, \\
        &= \int_{-\infty}^{+\infty} \Big[\int_{-\infty}^{+\infty} g(x) e^{-iux} \, \mathrm{d}x\Big] \Big[\int_{-\infty}^{+\infty} g(x') e^{-i(k-u)x'} \, \mathrm{d}x'\Big] \, \mathrm{d}u, \\
        &= \int_{-\infty}^{+\infty} \int_{-\infty}^{+\infty} \int_{-\infty}^{+\infty} g(x) g(x') e^{-iux} e^{iux'} e^{-ikx'} \, \mathrm{d}x \, \mathrm{d}x' \, \mathrm{d}u. \\
\end{align*}

Combine $e^{-iux}e^{iux'}$ into $e^{-i(x-x')u}$, we have:

\begin{align*}
    [F \star F](k) &= \int_{-\infty}^{+\infty} \int_{-\infty}^{+\infty} \int_{-\infty}^{+\infty} g(x) g(x') e^{-iu(x-x')} e^{-ikx'} \, \mathrm{d}x \, \mathrm{d}x' \, \mathrm{d}u, \\
        &= \int_{-\infty}^{+\infty} \int_{-\infty}^{+\infty} \Big[\int_{-\infty}^{+\infty} e^{-i(x-x')u} \, \mathrm{d}u\Big] g(x) g(x') e^{-ikx'} \, \mathrm{d}x \, \mathrm{d}x', \\
        &= \int_{-\infty}^{+\infty} \int_{-\infty}^{+\infty} \delta(x-x') g(x) g(x') e^{-ikx'} \, \mathrm{d}x \, \mathrm{d}x. \\
\end{align*}

Notice that $\delta(x-x')$ results in $x'$ being replaced by $x$, we have:

\begin{align*}
    [F \star F](k) &= \int_{-\infty}^{+\infty} \int_{-\infty}^{+\infty} \delta(x-x') g(x) g(x') e^{-ikx'} \, \mathrm{d}x \, \mathrm{d}x, \\
        &= \int_{-\infty}^{+\infty} g(x) g(x) e^{-ikx} \, \mathrm{d}x, \\
        &= \mathcal{F}[g(x)g(x)](k).  \tag{1}
\end{align*}

The Fourier transform can also be used to remove noise. The following illustration shows an image of the dog "Bailey" containing a noise signal and the corresponding Fourier spectrum.

`(a)` Describe what kind of noise signal is present above and in which way the frequency spectrum could be modified in order to eliminate the noise signal.

**Answer:**

The noise signal consists of two diagonal components and is periodic. This is reflected in the Fourier spectrum as four bright spots symmetrically positioned along the diagonal directions. To eliminate the noise, we can apply a  notch filter to suppress these specific frequencies.

`(b)` Write a Python script that removes the noise from the image `Noisy_Bailey.jpg`. 

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

image = np.array(Image.open('Noisy_Bailey.jpg').convert('L'))

rows, cols = image.shape
crow, ccol = rows // 2, cols // 2
notch_radius = 5
notch_pos = 25

notch_coords = [
    (crow - notch_pos, ccol - notch_pos),
    (crow + notch_pos, ccol + notch_pos),
    (crow - notch_pos, ccol + notch_pos),
    (crow + notch_pos, ccol - notch_pos)
]

notch_shifts = [
    [130, 0],
    [-130, 0],
    [0, 130],
    [0, -130],
]

def fft(image):
    f_transform = np.fft.fft2(image)
    f_shifted = np.fft.fftshift(f_transform)

    magnitude_spectrum = np.log(1 + np.abs(f_shifted))
    magnitude_spectrum = (magnitude_spectrum - np.min(magnitude_spectrum)) \
                       / (np.max(magnitude_spectrum) - np.min(magnitude_spectrum))

    return f_shifted, magnitude_spectrum

def ifft(f_shifted):
    f_inverse_shifted = np.fft.ifftshift(f_shifted)
    image_filtered = np.abs(np.fft.ifft2(f_inverse_shifted))

    magnitude_spectrum = np.log(1 + np.abs(f_shifted))
    magnitude_spectrum = (magnitude_spectrum - np.min(magnitude_spectrum)) \
                       / (np.max(magnitude_spectrum) - np.min(magnitude_spectrum))

    return image_filtered, magnitude_spectrum

def notch_filter(f_shifted, notch_coords, notch_radius):
    rows, cols = f_shifted.shape
    for x, y in notch_coords:
        for i in range(-notch_radius, notch_radius + 1):
            for j in range(-notch_radius, notch_radius + 1):
                if i**2 + j**2 <= notch_radius**2:
                    xi, yj = x + i, y + j
                    if 0 <= xi < rows and 0 <= yj < cols:
                        f_shifted[xi, yj] = 0

    return f_shifted

def notch_filter_shifted(f_shifted, notch_coords, notch_radius, shift, scale=2):
    for shift in notch_shifts:
        for i in range(scale):
            notch_coords_shifted = [(x + shift[0] * (i+1), y + shift[1] * (i+1)) for x, y in notch_coords]
            f_shifted = notch_filter(f_shifted, notch_coords_shifted, notch_radius)
    
    return f_shifted

f_shifted, magnitude_spectrum = fft(image)

f_shifted = notch_filter(f_shifted, notch_coords, notch_radius)
f_shifted = notch_filter_shifted(f_shifted, notch_coords, notch_radius, notch_shifts)

image_filtered, magnitude_spectrum_filtered = ifft(f_shifted)

plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Fourier Transform Spectrum')

plt.subplot(2, 2, 3)
plt.imshow(magnitude_spectrum_filtered, cmap='gray')
plt.title('Filtered Fourier Spectrum')

plt.subplot(2, 2, 4)
plt.imshow(image_filtered, cmap='gray')
plt.title('Filtered Image')

plt.show()
```
