The following cell provides a Python cell to show the installed packages (on your system or on the provided Jupyter Hub for Teaching).

Just run the cell below to display the installed packages. If you use your own Python installation and environment, install and use the packages provided in the Jupyter Hub for Teaching (https://jupyter-test.scc.kit.edu/).

```python
import subprocess
import pandas as pd
import matplotlib.pyplot as plt

# Retrieve the list of packages installed via pip
pip_output = subprocess.run(['pip3', 'list'], capture_output=True, text=True)
packages = pip_output.stdout

# Process the output into a pandas DataFrame
package_lines = packages.split('\n')[2:]  # Skip the first two lines
package_data = [line.split() for line in package_lines if line]  # Split by space and remove empty lines


# Set pandas to display all rows
pd.set_option('display.max_rows', None)

# Create a DataFrame with the package name and version
df = pd.DataFrame(package_data, columns=["Package", "Version"])

# Plot the data (e.g., bar plot)
df
```
