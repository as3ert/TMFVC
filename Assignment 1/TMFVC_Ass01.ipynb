### Exercise 1.1 (6 points)

Let $\mathbb{A}=\{(x,y)\in \mathbb{R}^2|x^2+y^2=1\}$ and $\vec{\mathbb{V}}=\vec{\mathbb{R}}$. The composition operation $\oplus:\mathbb{A}\times\vec{\mathbb{V}}\mapsto\mathbb{A}$, $(a,\vec{v})\mapsto a\oplus\vec{v}$ reads
	$$(x,y)\oplus(\alpha)=(x\cos\alpha-y\sin\alpha, x\sin\alpha+y\cos\alpha).$$
Which of the conditions of an affine space are satisfied by $(\mathbb{A},\vec{\mathbb{V}},\oplus)$? 
Please prove your answer for each of the conditions, respectively. 
Is $(\mathbb{A},\vec{\mathbb{V}},\oplus)$ an affine space?

1. For every $a\in \mathbb{A}$: $a \oplus \vec{0}=a$ is satisfied.

   For $\vec{0}\in\vec{\mathbb{V}}$, we have $\alpha=0$.

   Then, $(x,y) \oplus (0)=(x\cos 0-y\sin 0,x\sin 0+y\cos 0)$.
   
   We have $\sin(0)=0$ and $\cos(0)=1$.

   Thus, $(x,y) \oplus (0)=(x\cdot 1-y\cdot 0,x\cdot 0+y\cdot 1)=(x,y)$.

2. For any ordered pair $(p,q)$, with $p,q\in\mathbb{A}$, there is a unique $\vec{v}=\overrightarrow{pq}\in\vec{\mathbb{V}}$ such that $p\oplus\vec{v}=q$ is satisfied.

   Suppose we have two points $p,q\in\mathbb{A}$, in which $p=(x_p,y_p)$ and $q=(x_q,y_q)$.

   To find a unique $\vec{v}=\overrightarrow{pq}\in\vec{\mathbb{V}}$ such that $p\oplus\vec{v}=q$, which means find a unique $\alpha$.
   
   We have then some $\alpha$ which satisfies $(x_p,y_p)\oplus (\alpha)=(x_p\cos\alpha-y_p\sin\alpha,x_p\sin\alpha+y_p\cos\alpha)=(x_q,y_q)$,

   in which

   $$
       x_p\cos\alpha-y_p\sin\alpha=x_q \tag{1}
   $$
   $$
       x_p\sin\alpha+y_p\cos\alpha=y_q \tag{2}
   $$.

   First, we need to ensure that Eq. 1 and Eq. 2 holds at the same time. We can square both of the equations and add them, then we have

   $$
       x_p^2\cos^2\alpha+y_p^2\sin^2\alpha+x_p^2\sin^2\alpha+y_p^2\cos^2\alpha =x_q^2+y_q^2 \tag{3}
   $$.

   Recall that we also have

   $$
       x_p^2+y_p^2 = 1 
   $$
   $$
       x_q^2+y_q^2 = 1 
   $$, and

   $$
       \cos^2\alpha + \sin^2\alpha = 1.
   $$

   Thus, Eq. 3 turns out to be

   $$
   \begin{align}
       (x_p^2+y_p^2)\cos^2\alpha+(x_p^2+y_p^2)\sin^2\alpha &= x_q^2+y_q^2 \\
       \cos^2\alpha + \sin^2\alpha &= 1 \\
       1 &= 1
   \end{align}
   $$.

   Hence, here exists some $\alpha$ such that $(x_p,y_p)\oplus (\alpha)=(x_p\cos\alpha-y_p\sin\alpha,x_p\sin\alpha+y_p\cos\alpha)=(x_q,y_q)$.

   However, we still need to find the unique $\alpha$.

   Multiply Eq. 1 with $y_p$ and Eq. 2 with $x_p$, we have

   $$
       x_py_p\cos\alpha-y_p^2\sin\alpha=x_qy_p \tag{4}
   $$
   $$
       x_p^2\sin\alpha+y_px_p\cos\alpha=y_qx_p \tag{5}
   $$.

   Substract Eq. 4 from Eq. 5, we have

   $$
   \begin{align}
       (x_p^2+y_p^2)\sin\alpha &= y_qx_p-x_qy_p \\
       \sin\alpha &= y_qx_p-x_qy_p
   \end{align}
   $$.

   Similarly, multiply Eq. 1 with $x_p$ and Eq. 2 with $y_p$, we have

   $$
       x_p^2\cos\alpha-y_px_p\sin\alpha=x_qx_p \tag{6}
   $$
   $$
       x_py_p\sin\alpha+y_p^2\cos\alpha=y_qy_p \tag{7}
   $$.

   Add Eq. 6 to Eq. 7, we have

   $$
   \begin{align}
       (x_p^2+y_p^2)\cos\alpha &= x_qx_p+y_qy_p \\
       \cos\alpha &= x_qx_p+y_qy_p
   \end{align}
   $$.

   Thus, we have:
   
   $$
       \tan\alpha=\frac{y_qx_p-x_qy_p}{x_qx_p+y_qy_p}
   $$, which gives
   $$
       \alpha=\arctan\frac{y_qx_p-x_qy_p}{x_qx_p+y_qy_p}
   $$.

   So we find a unique $\alpha$ such that $(x_p,y_p)\oplus (\alpha)=(x_p\cos\alpha-y_p\sin\alpha,x_p\sin\alpha+y_p\cos\alpha)=(x_q,y_q)$ holds.

3. For every $a\in\mathbb{A}$ and every $\vec{u},\vec{v}\in\vec{\mathbb{V}}$: $(a\oplus\vec{u})\oplus\vec{v}=a\oplus(\vec{u}+\vec{v})$ is satisfied.

   Suppose we have two angles $\alpha, \beta\in\mathbb{R}$ and $(x,y)\in\mathbb{A}$.

   For the left-hand side (LHS) of the equation, we have:

   $$
   \begin{align}
       (x,y)\oplus\alpha\oplus\beta &= (x\cos\alpha-y\sin\alpha,x\sin\alpha+y\cos\alpha)\oplus\beta \\
                                    &= ((x\cos\alpha-y\sin\alpha)\cos\beta-(x\sin\alpha+y\cos\alpha)\sin\beta,(x\cos\alpha-y\sin\alpha)\sin\beta+(x\sin\alpha+y\cos\alpha)\cos\beta) \\
                                    &= (x\cos(\alpha+\beta)-y\sin(\alpha+\beta),x\sin(\alpha+\beta)+y\cos(\alpha+\beta))
   \end{align}
   $$

   And for the right-hand side (RHS), we have:

   $$
       (x,y)\oplus(\alpha+\beta) = (x\cos(\alpha+\beta)-y\sin(\alpha+\beta),x\sin(\alpha+\beta)+y\cos(\alpha+\beta))
   $$

   As we can see, the expressions for the LHS and the RHS are identical. Thus, we can show that
   $$
       (x,y)\oplus(\alpha+\beta)=(x,y)\oplus\alpha\oplus\beta
   $$.

All three conditions for an affine space are stisfied. Therefore, $(\mathbb{A},\vec{\mathbb{V}},\oplus)$ is an affine space.

### Exercise 1.2 (4 points)

Let $(\mathbb{A}_1,\vec{\mathbb{V}}_1,\oplus)$ and $(\mathbb{A}_2,\vec{\mathbb{V}}_2,\oplus)$ being two affine spaces with $\mathbb{A}_1=\mathbb{A}_2=\mathbb{A}$, $\vec{\mathbb{V}_1}=\vec{\mathbb{V}_2}=\vec{\mathbb{V}}$. Furthermore, $\vec{v}_0\in\vec{\mathbb{V}}$ is fixed.

Show that $F:\mathbb{A}\to\mathbb{A}$, $F(a):=a\oplus\vec{v}_0$ is an affinity.

*Hint: use $f=id$, i.e. $f(\vec{v}_0)=\vec{v}_0$.*

We have $F(a\oplus\vec{v})=F(a)\oplus f(\vec{v})$.

For the left-hand side (LHS) of the equation, we can get:

$$
\begin{align}
    F(a+\oplus\vec{v}) &= a\oplus\vec{v}\oplus\vec{v_0} \\
                       &= a\oplus(\vec{v}+\vec{v_0})
\end{align}
$$

For the right-hand side (RHS) of the equation, we have:

$$
\begin{align}
    F(a)\oplus f(\vec{v}) &= a\oplus\vec{v_0}\oplus\vec{v} \\
                          &= a\oplus(\vec{v_0}+\vec{v})
\end{align}
$$

Thus, $F$ is an affine map.

In order to proof that $F$ is bijective, we need to proof that $F$ is both injective and surjective.

For injection, we need to show that $F(a_1)=F(a_2)$ with $a_1,a_2\in\mathbb{A}$ implies $a_1=a_2$.

We have:

$$
\begin{align}
    F(a_1) &= a_1\oplus\vec{v_0} \\
    F(a_2) &= a_2\oplus\vec{v_0}
\end{align}
$$.

Thus, $a_1=a_2$.

For surjection, we need to show that for any $a_2\in\mathbb{A}$, we can always find $a_1\in\mathbb{A}$ such that $F(a_1)=a_2$.

Let $a_1=a_2\oplus(-\vec{v_0})$, we have then

$$
\begin{align}
    F(a_1) &= (a_2\oplus(-\vec{v_0}))\oplus\vec{v_0} \\
           &= a_2\oplus(\vec{v_0}-\vec{v_0}) = a_2
\end{align}
$$.

Thus, For any $a_2\in\mathbb{A}$, we can always find $a_1=a_2\oplus(-\vec{v_0})$ such that $F(a_1)=a_2$.

Since $F$ is both injective and surjective, $F$ is a bijective affine map. Thus, $F$ is an affinity.

### Exercise 1.3 (3 points)
Given is a triangle with vertices at $A=(0,0)$, $B=(3,1)$, and $C=(4,2)$. 

Create a Python script with functions to perform scaling and rotation on the triangleâ€™s vertices. Use the defined functions to scale the triangle with a factor of 2 and rotate it by 45 degrees.

Use `matplotlib` to plot the original triangle and its transformed versions on the same graph.

```python
import matplotlib.pyplot as plt
import numpy as np

# Store the three vertices of the trianngle
A = np.array([0, 0])
B = np.array([3, 1])
C = np.array([4, 2])

# Store the vertices in a matrix
vertices = np.array([A, B, C])

# Apply scaling transformation on the triangle
def scale(vertices, factor):
    scaling_matrix = np.array([[factor, 0], 
                               [0, factor]])
    scaled_vertices = np.dot(vertices, scaling_matrix)

    return scaled_vertices

# Apply rotation transformation on the triangle
def rotate(vertices, angle):
    rotation_matrix = np.array([[np.cos(angle), -np.sin(angle)], 
                                [np.sin(angle), np.cos(angle)]])
    rotated_vertices = np.dot(vertices, rotation_matrix)

    return rotated_vertices

# Scale the triangle by a factor of 2
scaled_vertices = scale(vertices, 2)

# Rotate the triangle by 45 degrees
rotated_vertices = rotate(vertices, np.pi/4)

# Plot the original triangle
plt.plot(np.append(vertices[:, 0], vertices[0, 0]), 
         np.append(vertices[:, 1], vertices[0, 1]), 'r', linewidth=2.5)

# Plot the scaled triangle
plt.plot(np.append(scaled_vertices[:, 0], scaled_vertices[0, 0]), 
         np.append(scaled_vertices[:, 1], scaled_vertices[0, 1]), 'g')

# Plot the rotated triangle
plt.plot(np.append(rotated_vertices[:, 0], rotated_vertices[0, 0]), 
         np.append(rotated_vertices[:, 1], rotated_vertices[0, 1]), 'b')

plt.show()
```
